### Data Description:
The dataset is sneakers.csv and it has anumber of 8 columns which are described below:

- Prodect Name: The name of the sneakers with brand and model.

- Sale Prices: The actual sale price  of the sneakers.
- Discount: The discount percentage.
- Brand: The sneakers' brand name.
- Rating: The rating of each product from custumer review.
- Review: Feedback from the customers.
