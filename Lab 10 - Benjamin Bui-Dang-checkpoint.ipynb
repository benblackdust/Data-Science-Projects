## Types of EDA:
1. **Univariate EDA**: We examine one feature (column, variable) in a univariate EDA.
2. **Bivariate EDA**: We examine two features (columns, variables) in a bivariate EDA.
3. **Multivariate**: We examine more than two features (columns, variables) in a mulitvariate EDA.
